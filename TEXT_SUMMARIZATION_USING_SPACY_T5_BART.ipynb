```python
import spacy
from string import punctuation
from collections  import Counter
```

```python
nlp=spacy.load('en_core_web_sm') #Loaded the english language Model to create a spcacy pipeline for text summarization, it does tokenization, NER
nlp
```

```python
text = """Sure! Here’s a large text on the topic of **Artificial Intelligence (AI):**

---

### **The Rise of Artificial Intelligence**
Artificial Intelligence (AI) has emerged as one of the most transformative technologies of the 21st century. Defined as the simulation of human intelligence in machines, AI encompasses a variety of subfields, including machine learning, natural language processing, computer vision, and robotics. The rapid advancements in these areas have led to AI applications becoming integral to many aspects of modern life.

One of the most visible impacts of AI is in automation. Industries such as manufacturing, logistics, and even customer service are leveraging AI to streamline operations. Robots powered by AI can perform repetitive tasks with precision, reducing human error and increasing efficiency. For example, companies like Amazon use AI-driven robots to manage inventory in warehouses, ensuring fast and accurate delivery of products. Similarly, self-driving cars, equipped with advanced AI systems, are set to revolutionize transportation by reducing accidents and optimizing traffic flow.

AI is also transforming healthcare. Machine learning algorithms are being used to analyze medical images, detect diseases, and predict patient outcomes. For instance, AI tools have demonstrated remarkable accuracy in identifying early signs of conditions like cancer from diagnostic scans. Additionally, AI-powered virtual assistants are helping patients manage chronic diseases, reminding them to take medications, and providing personalized health advice.

The education sector is witnessing a shift as well, with AI enabling personalized learning experiences. Adaptive learning platforms analyze student performance to tailor educational content, ensuring that each learner progresses at their own pace. These platforms can identify areas where a student struggles and provide targeted resources to improve understanding.

However, the rise of AI is not without challenges. Ethical concerns surrounding AI deployment are increasingly coming to the forefront. Issues such as data privacy, algorithmic bias, and the potential for job displacement raise questions about how society can responsibly integrate AI. For example, facial recognition technology, while useful for security purposes, has sparked debates over surveillance and the violation of individual rights. Furthermore, the "black box" nature of some AI systems, where decision-making processes are not easily interpretable, complicates accountability.

Despite these challenges, the potential benefits of AI are immense. Researchers are exploring AI-driven solutions for climate change, from optimizing renewable energy systems to monitoring deforestation using satellite imagery. In agriculture, AI is improving crop yields by analyzing soil conditions and predicting weather patterns.

Governments and organizations worldwide are investing heavily in AI research and development. Initiatives to promote AI literacy and ethical AI practices are gaining momentum. For instance, guidelines for responsible AI use are being drafted to ensure that technology serves humanity without causing harm.

In conclusion, AI is reshaping industries, improving lives, and driving innovation. While challenges remain, its potential to address global issues and create new opportunities is undeniable. As society navigates the complexities of AI adoption, striking a balance between innovation and responsibility will be key to unlocking its full potential.

---

Let me know if you'd like to summarize this or analyze it further!"""
```

```python
doc=nlp(text)
print([token.text for token in doc])
print()
print([(ent.text,ent.label_) for ent in doc.ents])
```

```python
print([token.text.lower() for token in doc])
```

```python
words = [token.text.lower() for token in doc if token.is_alpha and not token.is_stop]
words
```

```python
freq=Counter(words)
freq
```

```python
#Normalization to make them comparable and make their score fall between 0
for i in freq:
  freq[i]/=max(freq.values())
freq
```

```python
max_freq = max(freq.values())
for word in freq:
    freq[word] /= max_freq
freq
```

```python
sent_score={}
for sent in doc.sents:
  for word in sent:
    if word.text.lower() in freq.keys():
      if sent not in sent_score.keys():
        sent_score[sent]=freq[word.text.lower()]
sent_score
```

```python
top_sentences = sorted(sent_score, key=sent_score.get, reverse=True)[:5]
top_sentences
```

```python
summary=' '.join([sent.text for sent in top_sentences])
summary
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
tokenizer = T5Tokenizer.from_pretrained('t5-base')
model = T5ForConditionalGeneration.from_pretrained('t5-base')
```

```python
text = """Sure! Here’s a large text on the topic of **Artificial Intelligence (AI):**

---

### **The Rise of Artificial Intelligence**
Artificial Intelligence (AI) has emerged as one of the most transformative technologies of the 21st century. Defined as the simulation of human intelligence in machines, AI encompasses a variety of subfields, including machine learning, natural language processing, computer vision, and robotics. The rapid advancements in these areas have led to AI applications becoming integral to many aspects of modern life.

One of the most visible impacts of AI is in automation. Industries such as manufacturing, logistics, and even customer service are leveraging AI to streamline operations. Robots powered by AI can perform repetitive tasks with precision, reducing human error and increasing efficiency. For example, companies like Amazon use AI-driven robots to manage inventory in warehouses, ensuring fast and accurate delivery of products. Similarly, self-driving cars, equipped with advanced AI systems, are set to revolutionize transportation by reducing accidents and optimizing traffic flow.

AI is also transforming healthcare. Machine learning algorithms are being used to analyze medical images, detect diseases, and predict patient outcomes. For instance, AI tools have demonstrated remarkable accuracy in identifying early signs of conditions like cancer from diagnostic scans. Additionally, AI-powered virtual assistants are helping patients manage chronic diseases, reminding them to take medications, and providing personalized health advice.

The education sector is witnessing a shift as well, with AI enabling personalized learning experiences. Adaptive learning platforms analyze student performance to tailor educational content, ensuring that each learner progresses at their own pace. These platforms can identify areas where a student struggles and provide targeted resources to improve understanding.

However, the rise of AI is not without challenges. Ethical concerns surrounding AI deployment are increasingly coming to the forefront. Issues such as data privacy, algorithmic bias, and the potential for job displacement raise questions about how society can responsibly integrate AI. For example, facial recognition technology, while useful for security purposes, has sparked debates over surveillance and the violation of individual rights. Furthermore, the "black box" nature of some AI systems, where decision-making processes are not easily interpretable, complicates accountability.

Despite these challenges, the potential benefits of AI are immense. Researchers are exploring AI-driven solutions for climate change, from optimizing renewable energy systems to monitoring deforestation using satellite imagery. In agriculture, AI is improving crop yields by analyzing soil conditions and predicting weather patterns.

Governments and organizations worldwide are investing heavily in AI research and development. Initiatives to promote AI literacy and ethical AI practices are gaining momentum. For instance, guidelines for responsible AI use are being drafted to ensure that technology serves humanity without causing harm.

In conclusion, AI is reshaping industries, improving lives, and driving innovation. While challenges remain, its potential to address global issues and create new opportunities is undeniable. As society navigates the complexities of AI adoption, striking a balance between innovation and responsibility will be key to unlocking its full potential.

---

Let me know if you'd like to summarize this or analyze it further!"""
```

```python
inputs = tokenizer.encode("summarize: "+text, return_tensors="pt", max_length=512, truncation=True)
```

```python
outputs = model.generate(inputs, max_length=100, min_length=20, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
summary
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration
tokenizer=BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model=BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
```

```python
inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
```

```python
outputs = model.generate(inputs, max_length=50, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
summary
```
